## Data Cleaning in Pandas

```python
import pandas as pd
```

```python
df = pd.read_excel(r"C:\Users\Barno Bayramova\OneDrive\Рабочий стол\python practice\Data_Cleaning_in_Pandas\Customer Call List.xlsx")
df = df.drop_duplicates()
df
```

```python
import re

# Clean and format only valid 10-digit numbers
df.loc[:, "Phone_Number"] = df["Phone_Number"].apply(
    lambda x: re.sub(r"^(\d{3})(\d{3})(\d{4})$", r"\1-\2-\3", re.sub(r"\D", "", str(x)))
    if len(re.sub(r"\D", "", str(x))) == 10 else x
)
```

```python
df
```

```python
df["Phone_Number"]=df["Phone_Number"].str.replace('N/a','')
df["Phone_Number"] = df["Phone_Number"].fillna('')
df["Phone_Number"]
```

```python
df
```

```python
df[['Street', 'State', 'Zip']] = df['Address'].str.split(',', expand=True)
df
```

```python
df = df.drop(['Address', 'Not_Useful_Column'], axis=1)
```

```python
df
```

```python
df["Last_Name"] = df["Last_Name"].str.strip("123._/")
df
```

```python
df["Paying Customer"] = df["Paying Customer"].str.replace('Yes', 'Y')
df["Paying Customer"] = df["Paying Customer"].str.replace('No', 'N')
df
```

```python
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('Yes', 'Y')
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('No', 'N')
df
```

```python
df=df.replace('N/a','')

df=df.fillna('')
df
```

```python
for x in df.index:
    if df.loc[x,"Do_Not_Contact"] == 'Y':
        df.drop(x, inplace=True)
    
```

```python
df
```

```python
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('NNN', 'N')
```

```python
for x in df.index:
    if df.loc[x,"Phone_Number"] == '':
        df.drop(x, inplace=True)
    
```
