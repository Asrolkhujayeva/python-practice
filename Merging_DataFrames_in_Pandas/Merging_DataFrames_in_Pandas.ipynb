## Merge, Join and Concatinate

```python
import pandas as pd
```

```python
df1 = pd.read_csv(r"C:\Users\Barno Bayramova\OneDrive\Рабочий стол\python practice\Merging_DataFrames_in_Pandas\LOTR.csv")
df1
```

```python
df2 = pd.read_csv(r"C:\Users\Barno Bayramova\OneDrive\Рабочий стол\python practice\Merging_DataFrames_in_Pandas\LOTR 2.csv")
df2
```

```python
# with merge()
df1.merge(df2) # default inner join
```

```python
df1.merge(df2, how = 'inner', on = 'FellowshipID') # inner join
```

```python
df1.merge(df2, how = 'inner', on = ['FellowshipID', 'FirstName'])
```

```python
df1.merge(df2, how = 'outer')
```

```python
df1.merge(df2, how = 'left')
```

```python
df1.merge(df2, how = 'right')
```

```python
df1.merge(df2, how = 'cross')
```

```python
# with join(), it is used for index based joining
df1.join(df2, on = 'FellowshipID', how = 'outer', lsuffix = '_left', rsuffix = '_right')
```

```python
df4 = df1.set_index('FellowshipID').join(df2.set_index('FellowshipID'), how = 'outer', lsuffix = '_left', rsuffix = '_right')
df4
```

```python
# concatination, puts one after another
pd.concat([df1,df2,df4])
```

```python
# join with concat()
pd.concat([df1,df2], join = 'outer')
```

```python
pd.concat([df1,df2], join = 'outer', axis = 1)
```
